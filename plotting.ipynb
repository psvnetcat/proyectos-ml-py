```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
```

```python
wine = pd.read_csv('C:\\Users\\seba_\\Documents\\winemag-data_first150k.csv',index_col=0)
wine.head(3)
```

```python
(wine['country'].value_counts().head(12)/len(wine)).plot.bar()
```

```python
wine['points'].value_counts().sort_index().plot.bar()
```

```python
# grafica de cuantas veces los vinos gringos logran un determinado puntaje (ptje/frecuencia absoluta)
# sort_index() ordena el eje x
wine.loc[wine['country']=='US','points'].value_counts().sort_index().plot.bar()
```

```python
# grafica de cuantas veces los vinos gringos logran un determinado puntaje (ptje/frecuencia relativa) 
(wine.loc[wine['country']=='US','points'].value_counts()/len(wine)).sort_index().plot.bar()
```

```python
# si la escala de puntos pasara los cien, seria mejor plotearlas en linea (continua)
wine['points'].value_counts().sort_index().plot.line()
```
