```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import autocorrelation_plot, lag_plot
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import preprocessing
from statsmodels.tsa.ar_model import AR
```

```python
datosGas = pd.read_csv('consumo_gas.csv',encoding='latin-1', sep=';')
```

```python
datosGas['Fecha'] = pd.to_datetime(datosGas.apply(lambda x:'%s/%s/%s' % (x['Dia'],x['Mes'],x['anio']),axis=1))
```

```python
datosGas.head()
```

```python
serie = datosGas.loc[:,['Fecha','Consumo']]
```

```python
fig = plt.figure(figsize=(15,9))
ax = fig.add_subplot(111)
ax.plot_date(x=serie.Fecha, y=serie.Consumo, marker='o')
```

```python
plt.figure(figsize=(15,9))
lag_plot(serie.Consumo)
```

```python
plt.figure(figsize=(10,5))
autocorrelation_plot(serie['Consumo'])
```
